```python
import pandas as pd
df=pd.read_csv('/content/train.csv')
df
```

# 1. Identifying Missing Values

First, you need to identify the missing values in your dataset. You can use the following functions to do so:

```python
# # Displays the number of missing values in each column
print(df.isnull().sum())
```

```python
# Checks if there are any missing values in the columns
print(df.isnull().any())
```

# 2. Handling Missing Values

-There are several strategies to handle missing values, depending on the nature of your data and the problem you're solving:



```python
# Remove Rows with Missing Values: Use dropna() to remove rows containing missing values.
df_cleaned=df.dropna()
df_cleaned
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned.shape
```

```python
# Remove Columns with Missing Values: Use dropna(axis=1) to remove columns containing missing values.

df_cleaned=df.dropna(axis=1)
df_cleaned
```

# b. Replacing Missing Values

Fill with a Specific Value: Use fillna() to replace missing values with a specific value, such as 0 or the mean/median/mode of the column.



```python
# Fill missing values with 0
df_filled=df.fillna(0)
```

```python
df_filled.isnull().sum()
```

```python
df.info()
```

```python
numerical_cols=df.select_dtypes(include=['number'])
numerical_cols
```

```python
numerical_cols.info()
```

```python
categorical_cols=df.select_dtypes(include=['object'])
categorical_cols
```

```python
categorical_cols.info()
```

```python
categorical_cols1=df.select_dtypes(exclude=['number'])
categorical_cols1
```

```python
df[numerical_cols.columns]=numerical_cols.fillna(numerical_cols.mean())
df[numerical_cols.columns]
```

```python
df[categorical_cols.columns]=categorical_cols.apply(lambda x:x.fillna(x.mode()[0]))
df[categorical_cols.columns]
```

```python
df[categorical_cols.columns].isnull().sum()
```

# Forward/Backward Fill: Use ffill (forward fill) or bfill (backward fill) to propagate the next or previous values.

```python
# Forward fill
df_filled=df.fillna(method='ffill')
```

```python
df_filled.isnull().sum()
```

```python
df_filled1=df.fillna(method='bfill')
```

```python
df_filled1.isnull().sum()
```

```python
df_interpolated=df.interpolate()
df_interpolated
```

```python
df_interpolated.isnull().sum()
```

Imputation Using Scikit-Learn: Use the SimpleImputer or IterativeImputer from scikit-learn for more sophisticated imputation strategies.

```python
from sklearn.impute import SimpleImputer

# Impute with mean value
imputer = SimpleImputer(strategy='mean')
imputer_num=SimpleImputer(strategy='mean')
df[numerical_cols]=pd.DataFrame(imputer_num.fit_transform(df[numerical_cols]), columns=numerical_cols)
```
